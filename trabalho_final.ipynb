```python
import pandas as pd
import numpy as np

colunas_usadas = [
    "TP_FAIXA_ETARIA",
    "TP_SEXO",
    "TP_COR_RACA",
    "TP_NACIONALIDADE",
    "TP_ST_CONCLUSAO",
    "TP_ESCOLA",
    "IN_TREINEIRO",
    "TP_PRESENCA_CN",
    "TP_PRESENCA_CH",
    "TP_PRESENCA_LC",
    "TP_PRESENCA_MT",
    "TP_STATUS_REDACAO",
    "NU_NOTA_MT", 
    "NU_NOTA_REDACAO",
    "Q001",
    "Q002",
    "Q006",
]

mapeamento_raca = {
    0: "Não declarada",
    1: "Branca",
    2: "Preta",
    3: "Parda",
    4: "Amarela",
    5: "Indígena",
}

mapeamento_nacionalidade = {
    0: "Não informado",
    1: "Brasileiro(a)",
    2: "Brasileiro(a) Naturalizado(a)",
    3: "Estrangeiro(a)",
    4: "Brasileiro(a) Nato(a), nascido(a) no exterior",
}

mapeamento_Situaçao_de_Conclusao = {
    1: "Já concluí o Ensino Médio",
    2: "Estou cursando e concluirei o Ensino Médio em 2023",
    3: "Estou cursando e concluirei o Ensino Médio após 2023",
    4: "Não concluí e não estou cursando o Ensino Médio",
}

mapeamento_Faixa_etaria = {
    1: "Menor de 17 anos",
    2: "17 anos",
    3: "18 anos",
    4: "19 anos",
    5: "20 anos",
    6: "21 anos",
    7: "22 anos",
    8: "23 anos",
    9: "24 anos",
    10: "25 anos",
    11: "Entre 26 e 30 anos",
    12: "Entre 31 e 35 anos",
    13: "Entre 36 e 40 anos",
    14: "Entre 41 e 45 anos",
    15: "Entre 46 e 50 anos",
    16: "entre 51 e 55 anos",
    17: "Entre 56 e 60 anos",
    18: "Entre 61 e 65 anos",
    19: "Entre 66 e 70 anos",
    20: "Maior de 70 anos",
}

coluna_presenca = ["Presença CN", "Presença CH", "Presença LC", "Presença MT"]
mapeamento_tipo_escola = {
    1: "Nao respondeu",
    2: "Pública",
    3: "Privada",
}

mapeamento_treineiro = {
    1: "Sim",
    0: "Não",
}

mapeamento_presenca = {
    0: "Faltou a prova",
    1: "Presente na prova",
    2: "Eliminado na prova",
}

coluna_pais = ["Nivel academico do pai", "Nivel academico da mãe"]
mapeamento_academico_pais = {
    "A": "Nunca Estudou",
    "B": "NAO COMPLETOU A 4 SERIE/5 ANO DO ENSINO FUNDAMENTAL",
    "C": "COMPLETOU A 4 SERIE/5 ANO, MAS NAO COMPLETOU A 8 SERIE/9 ANO DO ENSINO FUNDAMENTAL",
    "D": "Completou a 8ª série/9º ano do Ensino Fundamental, mas não completou o Ensino Médio.",
    "E": "Completou o Ensino Médio, mas não completou a Faculdade.",
    "F": "Completou a Faculdade, mas não completou a Pós-graduação.",
    "G": "Completou a Pós-graduação.",
    "H": " Não sei.",
}

mapeamento_renda_familia = {
    "A": "NENHUMA RENDA",
    "B": "Até R$ 1.320,00",
    "C": "De R$ 1.320,01 até R$ 1.980,00.",
    "D": "De R$ 1.980,01 até R$ 2.640,00.",
    "E": "De R$ 2.640,01 até R$ 3.300,00.",
    "F": "De R$ 3.300,01 até R$ 3.960,00.",
    "G": "De R$ 3.960,01 até R$ 5.280,00.",
    "H": "De R$ 5.280,01 até R$ 6.600,00.",
    "I": "De R$ 6.600,01 até R$ 7.920,00.",
    "J": "De R$ 7.920,01 até R$ 9240,00.",
    "K": "De R$ 9.240,01 até R$ 10.560,00.",
    "L": "De R$ 10.560,01 até R$ 11.880,00.",
    "M": "De R$ 11.880,01 até R$ 13.200,00.",
    "N": "De R$ 13.200,01 até R$ 15.840,00.",
    "O": "De R$ 15.840,01 até R$19.800,00.",
    "P": "De R$ 19.800,01 até R$ 26.400,00.",
    "Q": "Acima de R$ 26.400,00.",
}

mapeamento_status_redacao = {
    1: "Sem problemas",
    2: "Anulada",
    3: "Cópia Texto Motivador",
    4: "Em Branco",
    6: "Fuga ao Tema",
    7: "Não atendimento ao tipo textual",
    8: "Texto insuficiente",
    9: "Parte desconectada",
}
```

```python
df = pd.read_csv(
    "MICRODADOS_ENEM_2023.csv", sep=";", usecols=colunas_usadas, encoding="latin-1"
)

df = df.rename(
    columns={
        "TP_FAIXA_ETARIA": "Faixa Etária",
        "TP_SEXO": "Sexo",
        "TP_COR_RACA": "Cor ou Raça",
        "TP_NACIONALIDADE": "Nacionalidade",
        "TP_ST_CONCLUSAO": "Situação de Conclusão",
        "TP_ESCOLA": "Tipo de Escola",
        "IN_TREINEIRO": "Treineiro",
        "TP_PRESENCA_CN": "Presença CN",
        "TP_PRESENCA_CH": "Presença CH",
        "TP_PRESENCA_LC": "Presença LC",
        "TP_PRESENCA_MT": "Presença MT",
        "TP_STATUS_REDACAO": "Status Redação",
        "Q001": "Nivel academico do pai",
        "Q002": "Nivel academico da mãe",
        "Q006": "Renda mensal da familia",
        "NU_NOTA_MT" : "Nota matematica", 
        "NU_NOTA_REDACAO" : "Nota Redação",
    }
)

df["Status Redação"] = df["Status Redação"].map(mapeamento_status_redacao)
df["Status Redação"] = df["Status Redação"].fillna("Situação não informada")

df["Nota Redação"] = pd.to_numeric(df["Nota Redação"], errors="coerce")
df["Nota matematica"] = pd.to_numeric(df["Nota matematica"], errors="coerce")
df["Nota Redação (exibição)"] = df["Nota Redação"].fillna("Faltou")
df["Nota matematica (exibição)"] = df["Nota matematica"].fillna("Faltou")


df["Cor ou Raça"] = df["Cor ou Raça"].map(mapeamento_raca)
df["Nacionalidade"] = df["Nacionalidade"].map(mapeamento_nacionalidade)
df["Situação de Conclusão"] = df["Situação de Conclusão"].map(
    mapeamento_Situaçao_de_Conclusao
)
df["Faixa Etária"] = df["Faixa Etária"].map(mapeamento_Faixa_etaria)
df["Tipo de Escola"] = df["Tipo de Escola"].map(mapeamento_tipo_escola)
df["Treineiro"] = df["Treineiro"].map(mapeamento_treineiro)
for coluna in coluna_presenca:
    df[coluna] = df[coluna].map(mapeamento_presenca)
for coluna in coluna_pais:
    df[coluna] = df[coluna].map(mapeamento_academico_pais)
df["Renda mensal da familia"] = df["Renda mensal da familia"].map(
    mapeamento_renda_familia
)

df.head()
```

### Perguntas:

#### 1 - Qual o percentual de faltas dos cantidatos na disciplina de Matemática e Redação?

```python
total_mt = len(df)
faltas_mt = (df["Presença MT"] == "Faltou a prova").sum()
percentual_faltas_mt = (faltas_mt / total_mt) * 100

total_redacao = len(df)
faltas_redacao = (df["Status Redação"] == "Situação não informada").sum()
percentual_faltas_redacao = (faltas_redacao / total_redacao) * 100
```

```python
print(f"Faltas em Matemática: {percentual_faltas_mt:.2f}%")
print(f"Faltas em Redação: {percentual_faltas_redacao:.2f}%")
```

#### 2 - Se formos olhar para a distribuição das idades dos alunos, como estaria a distribuição dos alunos por faixas etárias?

```python
distribuicao_idade = df["Faixa Etária"].value_counts().sort_index()
print(distribuicao_idade)
```

### 3 - Quem tem as melhores notas em matemática e redação, candidatos do sexo masculino ou feminino?

```python
media_por_sexo_db = (
    df_db.groupby("Sexo")[["Nota matematica", "Nota Redação"]].mean().round(2)
)

print(media_por_sexo_db)
```

### 4 - Há alguma influência no tipo da escola (publica e privada) nas notas de matemática e redação?

```python
df_filtrado = df_db[
    df_db["Tipo de Escola"].isin(["Pública", "Privada", "Nao respondeu"])
]

media_por_escola = (
    df_filtrado.groupby("Tipo de Escola")[["Nota matematica", "Nota Redação"]]
    .mean()
    .round(2)
)

print(media_por_escola)
```

### 5 - Qual a influência da escolaridade do Pai nas notas de Matemática e Redação? 

```python
media_por_escolaridade_pai = (
    df_db.groupby("Nivel academico do pai")[["Nota matematica", "Nota Redação"]]
    .mean()
    .round(2)
)

print(media_por_escolaridade_pai)
```

### 6 - Qual a influência da escolaridade do Mãe nas notas de Matemática e Redação? 

```python
media_por_escolaridade_mae = (
    df_db.groupby("Nivel academico da mãe")[["Nota matematica", "Nota Redação"]]
    .mean()
    .round(2)
)

print(media_por_escolaridade_mae)
```

### 7 - Qual a influência da renda do Pai e da Mãe nas notas de Matemática e Redação? 

```python
media_por_renda = (
    df_db.groupby("Renda mensal da familia")[["Nota matematica", "Nota Redação"]]
    .mean()
    .round(2)
)

print(media_por_renda)
```
